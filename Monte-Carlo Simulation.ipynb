```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_mean = 170
rev_std = 20
num_simulations = 1000
```

```python
# Created 1000 simulated revenues using the normal distribution with mean and std predefined
rev = np.random.normal(rev_mean, rev_std, num_simulations)
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
rev.mean(), rev.std()
```

```python
# In general, cogs is around 60% of the revenue, with standard deviation being 10%
# We don't need to run 1000 simulations here because rev has already 1000 data points
cogs = - (rev * np.random.normal(0.6, 0.1)) 
plt.figure(figsize=(15,6))
plt.plot(cogs)
plt.show()
```

```python
cogs.mean(),  cogs.std()
```

```python
# Calculate gross profit
gross_profit = rev + cogs
plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```

```python
gross_profit.mean(), gross_profit.std()
```

```python
# plot the histogram of the gross profit
plt.figure(figsize=(10,6))
plt.hist(gross_profit, bins=20)
plt.show()
```
