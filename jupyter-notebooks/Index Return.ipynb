```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI']
data = pd.DataFrame()
for ticker in tickers:
    data[ticker] = wb.DataReader(ticker, data_source='yahoo', start='1995-1-1')['Adj Close']

data.head()
```

```python
# normalization to 100
(data / data.iloc[0] * 100).plot(figsize=(20,10))
```

```python
data.tail()
```

```python
returns = ((data / data.shift(1)) -1 ).dropna()
returns
```

```python
mean_return = returns.mean() * 252
mean_return
```
